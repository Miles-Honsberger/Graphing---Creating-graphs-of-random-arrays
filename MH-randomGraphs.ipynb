```python
#1 random graph generator
import random
def randomGraph(n, p): #n vertices, probability p
  adjacencies = [[] for _ in range(n)]
  for i in range(n): #i and j respectively represent the rows and columns of the adj list
    for j in range(i):
      if random.random() < p:
        adjacencies[i].append(j)
        adjacencies[j].append(i)
  return adjacencies
```

```python
randomGraph(50,0.5)
```

```python
#2 plot tool
import networkx as nx
import matplotlib.pyplot as plt

def adjListToGraph(adjacencyList):
  G = nx.Graph()
  number_nodes = len(adjacencyList) #n vertices
  nodes = range(number_nodes) # 0 to n-1
  G.add_nodes_from(nodes)
  for i, neighbours in enumerate(adjacencyList): 
    for neighbour in neighbours:
      G.add_edge(i, neighbour)
  return nx.draw(G)
```

```python
#3
def largestComp(adjList):
  largest = [] #This list will be the largest component
  for i in range(len(adjList)): #assigns a number value to i in the range of the adjacency list
    if len(adjList[i]) > len(largest): #if the subsequent component is larger than the previous, the subsequent component is the new largest
      largest = adjList[i] #at the start 'largest' will become the first component
  return len(largest) #returns the length of the largest component
```

```python
#4
def expectedLargest(n,p,trials):
  largest = []
  adjacencies = [[] for _ in range(n)] #the same code to generate a graph
  avgNormSize = 0
  for _ in range(trials): #repeats the experiment for each # of trials
    for i in range(n): # generates an adjacency list
      for j in range(i):
        if random.random() < p:
          adjacencies[i].append(j)
          adjacencies[j].append(i)
    
    for k in range(len(adjacencies)): #checks every component in the list
      if len(adjacencies[k]) > len(largest): #checks if the subsequent component is larger than the previous
        largest = adjacencies[k] #largest will be the largest component
    normSize = len(largest) / n #calculates the normalized size
    avgNormSize += normSize #sum of all normalized sizes
  return avgNormSize / trials #returns the average of the normalized sizes
```

```python
#5
'''
upon running the code with expectedLargest(20, 0.05, 1) we obtain 0.15 as the most common normalized size
'''
def expectedLargest2(n,t,increment):
  largest = []
  adjacencies = [[] for _ in range(n)] 
  t0 = 0 #represents the current value of t
  normSizes = [] #list for the normalised sizes of the largest component
  while t0 < t: #the code runs until t0 equals t
    for i in range(n): 
      for j in range(i):
        if random.random() < t0/n:
          adjacencies[i].append(j)
          adjacencies[j].append(i)
    t0 += increment #t0 is incrememnted after each graph is generated
    
    for component in adjacencies: 
      if len(component) > len(largest): 
        largest = component 
    normSizes.append(len(largest)/n) #normalized sizes are the length of the largest component / # of vertices
  return normSizes

def expLargPlotter(n,t,increment):
  largest = []
  adjacencies = [[] for _ in range(n)] 
  t0 = 0
  x = [] #t values
  y = [] #normalized sizes
  while t0 < t:
    for i in range(n): 
      for j in range(i):
        if random.random() < t0/n:
          adjacencies[i].append(j)
          adjacencies[j].append(i)
    x.append(t0) #current t value is added to the x value list
    t0 += increment #t0 is incremented after appending
    
    for k in range(len(adjacencies)): 
      if len(adjacencies[k]) > len(largest): 
        largest = adjacencies[k] 
    y.append(len(largest)/n) #the normalized size value is appended after the largest has been found
  return plt.plot(x,y) #simple plot function
```

```python
#6 code to calculate the ratio with trials and plotting tools
def calcRatio(n,p,trials):
  largest = []
  largest2 = []
  adjacencies = [[] for _ in range(n)] #the code for generating the adjacency lists is identical to previous parts
  for _ in range(trials): 
    for i in range(n): 
      for j in range(i):
        if random.random() < p:
          adjacencies[i].append(j)
          adjacencies[j].append(i)
    
    for component in adjacencies: 
      if len(component) > len(largest): 
        largest = component #the largest component is found and assigned
      adjacencies.remove(component) #this component is removed
    for component2 in adjacencies: #the second largest component is found and assigned
      if len(component2) > len(largest2): 
        largest2 = component2
    ratio = len(largest) / len(largest2) #ratio value is NOT made to be an integer as most ratios are between 0 and 1.5
  return ratio / trials #keeping the return value to be a floating number gives a more accurate result

```

```python
def plotRatio(n,t,increment):
  largest = [] #longest component
  largest2 = [] #second longest component
  dupe = []
  adjacencies = [[] for _ in range(n)] 
  t0 = 0
  x = [] #holds the t values
  y = [] #holds the ratios
  while t0 < t: 
    for i in range(n): 
      for j in range(i):
        if random.random() < t0/n:
          adjacencies[i].append(j) 
          adjacencies[j].append(i) #code that generates the adjacency lists
    x.append(t0) #current t0 value is added to the x axis list
    t0 += increment #t0 is increased afterwards
    dupe = list(adjacencies) #a duplicate of the adjacency list is made to prevent index errors
    for comp in dupe: #searchs each component
      if len(comp) == max(len(k) for k in dupe): #finds the longest component
        largest = max(len(k) for k in dupe) #largest is the longest component
      dupe.remove(comp) #the largest is removed
      largest2 = max(len(com) for com in dupe) #the second largest is found
    if largest != 0: #prevents a divide by zero error is largest2 is 0
      y.append(largest/largest2)
    else:
      y.append(0) #if largest2 is 0 the ratio is 0
  return plt.plot(x,y)
```
